```python
#Reading the text file containting meta deta
fh = open('amazon-meta.txt','r')
# List that will contain all the scraped info
Ids = []
ASIN = []
title =[]
group =[]
salesrank = []
similar = []
categories = []
rating = []

#Scraping meta deta, cleaning, and adding it into the lists for storage
for i in fh.readlines():
    if "Id" == i[:2]:
        if i.isdigit:
            Ids.append((int(i[6:])))
    if "ASIN" == i[:4]:
        ASIN.append(i[6:].strip("\n"))
    if i.find("title") >= 0:
        title.append(i[i.find("title")+6:].strip("\n"))
    if i.find("group") >= 0:
        group.append(i[i.find("group") + 6:].strip("\n"))
    if i.find("salesrank") >= 0:
        salesrank.append(int(i[i.find("salesrank")+10:].strip("\n")))
    if i.find("similar") >= 0:
        similar.append(i[i.find("similar")+8:].strip("\n"))
    if i.find("categories") >= 0:
        categories.append(int(i[i.find("categories")+12:].strip("\n")))
    if i.find("avg rating") >= 0:
        rating.append(float(i[i.find("avg rating")+12:].strip("\n")))

#Temporarily creating a dictionary, to convert to pandas dataframe
amazon_database = {'ID':Ids[1:], 'ASIN':ASIN[1:], 'Title': title, "Group": group, 'Sales_Rank':salesrank,"Similar_Books":similar,"No_of_Categories":categories,"Rating":rating}
df = pd.DataFrame.from_dict(amazon_database)
print(df)
        

    
    
    
        
        
print(len(Ids))
print(len(ASIN))
print(len(title))
print(len(group))
print(len(salesrank))
print(len(similar))
print(len(categories))
print(len(rating))
fh.close()

                
            

                           
```
